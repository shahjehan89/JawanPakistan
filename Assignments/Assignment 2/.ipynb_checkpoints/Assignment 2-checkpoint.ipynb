```python
import pandas as pd

data = pd.read_csv("Pokemon.csv")

data.head(10)
```

```python
round(data["HP"].mean())
```

```python
hp_mean = round(data["HP"].mean())

def set_hp(val):
    if val > hp_mean:
        return "HP is High"
    elif val == hp_mean:
        return "HP is equal"
    else:
        return "HP is Low"
    
data["High_Low_HP"] = data["HP"].apply(set_hp)

data.head(510)
```

```python
speed_mean = round(data["Speed"].mean())

def set_speed(val):
    if val > speed_mean:
        return "Speed is High"
    elif val == speed_mean:
        return "Speed is equal"
    else:
        return "Speed is Low"
    
data["High_Low_Speed"] = data["Speed"].apply(set_speed)

data.head(130)
```
