# MATPLOTLIB AND NUMPY

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,10,1000)

# fig = plt.Figure()

plt.plot(x,np.sin(x))
plt.plot(x,np.cos(x))

# fig.savefig("LinePlot.png")

plt.show()
```

```python
plt.figure()

plt.subplot(3,1,1)
plt.plot(x,np.sin(x))

plt.subplot(3,1,2)
plt.plot(x,np.cos(x))

plt.subplot(3,1,3)
plt.plot(x,np.tan(x))

plt.show()
```

```python
# Object Orientent

fig, ax = plt.subplots(3)

ax[0].plot(x,np.sin(x))
ax[1].plot(x,np.cos(x))
ax[2].plot(x,np.tan(x))
```

```python
plt.style.use("seaborn-whitegrid")
```

```python
x = np.linspace(0,10,1000)

plt.plot(x , np.sin(x), color = 'y')
plt.plot(x, np.cos(x), color = "g")

plt.show()
```

```python
x = np.linspace(0,10,1000)

plt.plot(x , np.sin(x), color = 'y', linestyle="solid" , label = "sin(x)")
plt.plot(x, np.cos(x-1), color = "g", linestyle="dotted" , label = "cos(x-1)")
plt.plot(x , np.sin(x-2), color = 'r', linestyle="dashed" , label = "sin(x-2)")
plt.plot(x, np.cos(x-3), color = "b", linestyle="dashdot" , label = "cos(x-3)")

# Adding Labels
plt.title("Sign Graph")
plt.xlabel("x")
plt.ylabel("sin(x)")

plt.legend()
plt.show()
```
